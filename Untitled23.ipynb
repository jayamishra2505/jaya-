```python
import pandas as pd
df = pd.read_csv(filename)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from pywaffle import Waffle
```

```python
df.head()
```

```python
team_runs = df.groupby('batting_team')['runs_off_bat'].sum().reset_index()

# Bar plot for run distribution by teams
plt.figure(figsize=(12, 6))
sns.barplot(x='batting_team', y='runs_off_bat', data=team_runs, palette='rocket')
plt.title('Total Runs Scored by Each Team')
plt.xlabel('Team')
plt.ylabel('Total Runs')
plt.xticks(rotation=90)
plt.show()
```

```python
# Calculate the sum of each type of extras
extras = df[['wides', 'noballs', 'byes', 'legbyes', 'penalty']].sum().reset_index()
extras.columns = ['extra_type', 'count']

# Bar plot for extras distribution
plt.figure(figsize=(12, 6))
sns.barplot(x='extra_type', y='count', data=extras, palette='mako')
plt.title('Distribution of Different Types of Extras')
plt.xlabel('Extra Type')
plt.ylabel('Count')
plt.show()
```

```python
# Calculate total runs per over
df['over'] = df['ball'].apply(lambda x: int(x))
runs_per_over = df.groupby('over')['runs_off_bat'].sum().reset_index()

# Line plot for runs per over
plt.figure(figsize=(12, 6))
sns.lineplot(x='over', y='runs_off_bat', data=runs_per_over, marker='o')
plt.title('Runs Scored per Over')
plt.xlabel('Over')
plt.ylabel('Total Runs')
plt.show()
```

```python
# Calculate total runs by each batsman
batsman_runs = df.groupby('striker')['runs_off_bat'].sum().reset_index()

# Get top 10 batsmen
top_batsmen = batsman_runs.sort_values(by='runs_off_bat', ascending=False).head(10)

# Bar plot for top batsmen runs
plt.figure(figsize=(12, 6))
sns.barplot(x='runs_off_bat', y='striker', data=top_batsmen, palette='viridis')
plt.title('Runs Distribution for Top 10 Batsmen')
plt.xlabel('Total Runs')
plt.ylabel('Batsman')
plt.show()
```

```python
# Pivot table for runs by over and team
runs_pivot = df.pivot_table(values='runs_off_bat', index='over', columns='batting_team', aggfunc='sum', fill_value=0)

# Heatmap for runs by over and team
plt.figure(figsize=(12, 8))
sns.heatmap(runs_pivot, cmap='viridis', annot=True, fmt='d')
plt.title('Heatmap of Runs by Over and Team')
plt.xlabel('Team')
plt.ylabel('Over')
plt.show()
```

```python
# Calculate total runs by venue
venue_runs = df.groupby('venue')['runs_off_bat'].sum().reset_index()

# Bar plot for total runs by venue
plt.figure(figsize=(12, 6))
sns.barplot(x='runs_off_bat', y='venue', data=venue_runs.sort_values(by='runs_off_bat', ascending=False), palette='cubehelix')
plt.title('Total Runs by Venue')
plt.xlabel('Total Runs')
plt.ylabel('Venue')
plt.show()
```

```python
# Calculate total wickets by each bowler
bowler_wickets = df[df['wicket_type'].notna()].groupby('bowler').size().reset_index(name='wickets')

# Get top 10 wicket takers
top_wicket_takers = bowler_wickets.sort_values(by='wickets', ascending=False).head(10)

# Bar plot for top wicket takers
plt.figure(figsize=(12, 6))
sns.barplot(x='wickets', y='bowler', data=top_wicket_takers, palette='coolwarm')
plt.title('Top 10 Wicket Takers')
plt.xlabel('Total Wickets')
plt.ylabel('Bowler')
plt.show()
```

```python
# Calculate total balls bowled by each bowler
balls_bowled = df.groupby('bowler').size().reset_index(name='balls_bowled')

# Calculate total runs conceded by each bowler
runs_conceded = df.groupby('bowler')['runs_off_bat'].sum().reset_index()

# Merge the data
bowler_stats = pd.merge(balls_bowled, runs_conceded, on='bowler')
bowler_stats['economy_rate'] = (bowler_stats['runs_off_bat'] / bowler_stats['balls_bowled']) * 6

# Get top 10 bowlers by economy rate (minimum 60 balls bowled)
top_economy_bowlers = bowler_stats[bowler_stats['balls_bowled'] >= 60].sort_values(by='economy_rate').head(10)

# Bar plot for economy rate of top bowlers
plt.figure(figsize=(12, 6))
sns.scatterplot(x='economy_rate', y='bowler', data=top_economy_bowlers)
plt.title('Top 10 Bowlers by Economy Rate')
plt.xlabel('Economy Rate')
plt.ylabel('Bowler')
plt.show()
```

```python
# Get top 10 bowlers by runs conceded
top_runs_conceded_bowlers = bowler_stats.sort_values(by='runs_off_bat', ascending=False).head(10)

# Bar plot for runs conceded by top bowlers
plt.figure(figsize=(12, 6))
sns.barplot(x='runs_off_bat', y='bowler', data=top_runs_conceded_bowlers, palette='pink')
plt.title('Top 10 Bowlers by Runs Conceded')
plt.xlabel('Total Runs Conceded')
plt.ylabel('Bowler')
plt.show()
```

```python
# Calculate total balls faced by each batsman
balls_faced = df.groupby('striker').size().reset_index(name='balls_faced')

# Merge with runs data
batsman_stats = pd.merge(batsman_runs, balls_faced, on='striker')
batsman_stats['strike_rate'] = (batsman_stats['runs_off_bat'] / batsman_stats['balls_faced']) * 100

# Get top 10 batsmen by strike rate
top_strike_rate_batsmen = batsman_stats.sort_values(by='strike_rate', ascending=False).head(10)

# Bar plot for strike rate of top batsmen
plt.figure(figsize=(12, 6))
sns.lineplot(x='strike_rate', y='striker', data=top_strike_rate_batsmen, palette='viridis')
plt.title('Top 10 Batsmen by Strike Rate')
plt.xlabel('Strike Rate')
plt.ylabel('Batsman')
plt.show()
```

```python
# Calculate total sixes hit by each batsman
sixes = df[df['runs_off_bat'] == 6].groupby('striker').size().reset_index(name='sixes')

# Get top 10 six hitters
top_six_hitters = sixes.sort_values(by='sixes', ascending=False).head(10)

# Bar plot for top six hitters
plt.figure(figsize=(12, 6))
sns.barplot(x='sixes', y='striker', data=top_six_hitters, palette='viridis')
plt.title('Top 10 Six Hitters')
plt.xlabel('Total Sixes')
plt.ylabel('Batsman')
plt.show()
```

```python
# Calculate total fours hit by each batsman
fours = df[df['runs_off_bat'] == 4].groupby('striker').size().reset_index(name='fours')

# Get top 10 four hitters
top_four_hitters = fours.sort_values(by='fours', ascending=False).head(10)

# Strip plot for top four hitters
plt.figure(figsize=(12, 6))
sns.stripplot(x='fours', y='striker', data=top_four_hitters, palette='mako', size=10)
plt.title('Top 10 Four Hitters')
plt.xlabel('Total Fours')
plt.ylabel('Batsman')
plt.show()

```

```python
# Calculate total dot balls faced by each batsman
dot_balls = df[df['runs_off_bat'] == 0].groupby('striker').size().reset_index(name='dot_balls')

# Get top 10 players who faced most dot balls
top_dot_ball_faced = dot_balls.sort_values(by='dot_balls', ascending=False).head(10)

# Point plot for most dot balls faced
plt.figure(figsize=(12, 6))
sns.pointplot(x='dot_balls', y='striker', data=top_dot_ball_faced, palette='cubehelix', join=False)
plt.title('Top 10 Players by Dot Balls Faced')
plt.xlabel('Total Dot Balls')
plt.ylabel('Batsman')
plt.show()
```
