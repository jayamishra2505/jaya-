```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np
import pandas as pd

```

```python
plt.figure(figsize=(10, 9))
sns.lineplot(x='Captain', y='Score', data=df, marker='o')
plt.title('Rohit Sharma Centuries Over the Years')
plt.xlabel('captaincy')
plt.ylabel('Number of Centuries')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(25, 10))
sns.barplot(x='Venue', y='Score', data=df, palette='viridis')
plt.title('Rohit Sharma Centuries Over the Years')
plt.xlabel('venue')
plt.ylabel('Number of Centuries')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Strike Rate'], df['Score'], color='skyblue')
plt.title('Rohit Sharma strike rate  (Bar Chart)')
plt.xlabel('strike rate')
plt.ylabel('Number of Centuries')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Score'], bins=10, color='orange', edgecolor='black')
plt.title('Histogram of Rohit Sharma Centuries')
plt.xlabel('Number of Centuries')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(df['Result'], color='brown')
plt.title('Violin Plot of rohit sharma matching winning or losing')
plt.xlabel('won/loss')
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(8, 6))
sns.boxplot(df['Position'], color='orange')
plt.title('Box Plot of Rohit Sharma position')
plt.xlabel('Number')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['Innings'], df['Score'], color='blue', edgecolor='k')
plt.title('Scatter Plot of Rohit Sharma Centuries Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Centuries')
plt.grid(True)
plt.show()
```
